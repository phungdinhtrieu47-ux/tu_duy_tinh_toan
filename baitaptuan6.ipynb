bai 1 : Nhập vào một dòng gồm các số nguyên. Hãy loại bỏ phần tử trùng nhưng
giữ nguyên thứ tự xuất hiện đầu tiên, rồi in ra list kết quả.
Yêu cầu in ra đúng dạng list Python, ví dụ [1, 2, 3]

```python
nums = list(map(int, input().split()))
result = []
seen = set()
for n in nums:
    if n not in seen:
        result.append(n)
        seen.add(n)
print(result)
```

bai 4 : Nhập một dòng gồm các cặp key:value cách nhau bởi dấu cách. Tạo 
dictionary ánh xạ mỗi key -> list các value theo thứ tự xuất hiện. Nếu key 
lặp lại, thêm vào list của key đó

```python
pairs = input().split()
d = {}
for p in pairs:
    key, value = p.split(":")
    if key not in d:
        d[key] = []
    d[key].append(value)

print(d)
```

bai 5 : Nhập một list số nguyên, In ra dictionary {'positives': ..., 'negatives': ..., 'zeros': ...} (đếm số >0, <0, =0) 

```python
lst = list(map(int, input().split()))
dis = {"positive": 0, "negative": 0, "zero": 0}
for n in lst:
    if n > 0:
        dis["positive"] += 1
    elif n < 0:
        dis["negative"] += 1
    else:
        dis["zero"] += 1
print(dis)
```

bai 9 : Viết chương trình nhập vào hai dictionary. Sau đó gộp lại thành một 
dictionary mới trong đó nếu một key xuất hiện trong cả hai dict, giá trị sẽ 
được cộng lại, nếu chỉ có trong một dict thì giữ nguyên. In ra dictionary mới 
theo thứ tự từ điển tăng dần của key.  

```python
s1 = input("Nhập dict1 (theo dạng: key value ...): ")
parts1 = s1.split()
dict1 = {parts1[i]: int(parts1[i+1]) for i in range(0, len(parts1), 2)}
s2 = input("Nhập dict2 (theo dạng: key value ...): ")
parts2 = s2.split()
dict2 = {parts2[i]: int(parts2[i+1]) for i in range(0, len(parts2), 2)}
merged_dict = dict1.copy()
for k, v in dict2.items():
    if k in merged_dict:
        merged_dict[k] += v
    else:
        merged_dict[k] = v
sorted_dict = {k: merged_dict[k] for k in sorted(merged_dict)}
print(sorted_dict)
```

Bai 10 : Viết chương trình nhập vào một danh sách các số nguyên, và số nguyên 
dương k. In ra list gồm các cặp số khác nhau có tổng bằng k. Mỗi cặp là 
một tuple, hai số trong tuple có thể giống nhau, thứ tự xuất hiện trong danh 
sách của số thứ nhất nhỏ hơn thứ tự của số thứ hai. Các tuple trong list được 
sắp xếp theo thứ tự tăng dần của phần tử đầu tiên trong tuple.

```python
nums = list(map(int, input("Nhập danh sách số nguyên (cách nhau bởi khoảng trắng): ").split()))
k = int(input("Nhập số nguyên dương k: "))
pairs = []
n = len(nums)
for i in range(n):
    for j in range(i+1, n):
        if nums[i] + nums[j] == k:
            pairs.append((nums[i], nums[j]))
pairs.sort(key=lambda x: x[0])
print(pairs)
```

bai 12 : [Tìm phần tử xuất hiện nhiều nhất] 
Nhập vào một list các số nguyên. Hãy tìm số xuất hiện nhiều lần nhất trong 
list. Nếu có nhiều số cùng có số lần xuất hiện nhiều nhất, hãy in ra số nhỏ 
nhất trong số đó. 

```python
lst = list(map(int,input().split()))
st = set(lst)
dic = {}
for x in st:
    dic[x] = lst.count(x)
a = max(dic.values())
lst1 = []
for x in st:
    if dic[x] == a:
        lst1.append(x)
print(min(lst1))
```

bai 13 : [Đảo ngược key và value của dictionary] 
Nhập vào một dictionary có value là duy nhất (không trùng nhau). Hãy tạo 
một dictionary mới bằng cách đảo ngược key và value của dictionary ban 
đầu. In ra dictionary mới.

```python
n = input()
s = n.split()
dis = {}
for i in range(0, len(s), 2):
    dis[s[i]] = s[i + 1]
dis1 = {v: k for k, v in dis.items()}
print(dis1)
```

bai 15 : [Lọc dictionary theo giá trị] 
Nhập vào một dictionary có key là chuỗi và value là số nguyên, và một số 
nguyên k ở dòng tiếp theo. Hãy tạo một dictionary mới chỉ chứa các cặp 
key-value từ dictionary ban đầu mà có value lớn hơn k. In ra dictionary 
mới.

```python
n = input()
s = n.split()
k = int(input())
dis = {}
for i in range(0, len(s), 2):
    dis[s[i]] = s[i + 1]
dis2 = {}
for x in dis:
    if dis[x] == str(k):
        dis2[x] = dis[x]
print(dis2)
    
```
