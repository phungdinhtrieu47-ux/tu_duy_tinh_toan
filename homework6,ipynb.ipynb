bai 16 : Nhậ ̣p vào mảng n hàng, m cột số nguyên. In lại mảng ra màn hình như ma trận hai chiều, định dạng mỗi phần tử 4 vị trí căn phải 

```python
n = int(input("Nhap so hang n: "))
m = int(input("Nhap so cot m: "))
a = []
print("Nhap cac phan tu cua mang:")
for i in range(n):
    row = []
    for j in range(m):
        x = int(input(f"a[{i}][{j}] = "))
        row.append(x)
    a.append(row)
print("Ma tran:")
for i in range(n):
    for j in range(m):
        print(f"{a[i][j]:4}", end="")
    print()
```

bai 17 : Nhập vào mảng vuông 2 chiều n hàng, n cột, in ra đường chéo chính và 
đường chéo phụ của mảng 

```python
n = int(input())
a =[]
for i in range(n):

    x = map(int, input().split())
    a.append(list(x))
for i in range(n):
    for j in range(n):
        if i == j:
            print(a[i][j], end=" ")
print()
for i in range(n):
    for j in range(n):
        if i + j == n - 1:
            print(a[i][j], end=" ")
```

bai 20 : Viết hàm đưa vào một list số nguyên và một số nguyên dương k. Hãy tìm và trả về vị trí của phần tử đầu tiên có giá trị k trong list, nếu không có thì trả về -1

```python
n = list(map(int, input().split()))
k = int(input())
for x, i in zip(n, range(len(n))):
    if x == k:
        print(i)
        break
if i == len(n) - 1 and n[i] != k:
    print(-1)
```

bai 26 : Tính tổng các phần tử cực đại địa phương trong một danh sách, không tính 2 phần tử đầu và cuối 

```python
n = list(map(int, input().split()))
x = 0
for i in range(1, len(n) - 1):
    if n[i] > n[i - 1] and n[i] > n[i + 1]:
        x += n[i]
print(x)
```

bai 27 : Cho một dãy số nguyên, Tìm và in ra vị trí các số < x (x là giá trị cho trước). nếu không có in ra -1

```python
n = list(map(int, input().split()))
x = int(input())
b = 0
for a in n:
    if a < x:
        print(n.index(a), end=" ")
        b += 1
if b == 0:
    print(-1)
```

bai 28 : Cho một dãy các số nguyên, kiểm tra xem nó là dãy tăng thực sự hay giảm thực sự, hay vô danh (không tăng, không giảm)

```python
n = list(map(int, input().split()))
a = 0
for i in range(len(n) - 1):
    if n[i] < n[i + 1]:
        a += 1
b = 0
for i in range(len(n) - 1):
    if n[i] > n[i + 1]:
        b += 1
if a == len(n) - 1:
    print("Tang dan")
elif b == len(n) - 1:
    print("Giam dan")
else:
    print("Khong tang giam")
    
```
