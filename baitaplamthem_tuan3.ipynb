bai 9 :  Nhập vào số nguyên dương a, kiểm tra xem a có phải là số chính phương hay 
không? Số chính phương là số có căn bậc hai là một số nguyên VD: 1, 4, 9, 16, 25….

```python
import math
a = int(input())
x = int(math.sqrt(a))
if x * x == a:
    print(a, "la so chinh phuong")
else:
    print(a, "khong phai la so chinh phuong") 
```

bai 10 : Nhập vào số nguyên dương a, kiểm tra xem a có phải là số hoàn hảo hay không. Số hoàn hảo là số có tổng các ước (ước nhỏ hơn nó) bằng chính nó, VD: số 6, 28… 

```python
tong = 0
a = int(input())
for i in range(1, a):
    if a % i == 0:
        tong += i
if tong == a:
    print(a, "la so hoan hao")
else:
    print(a, "khong phai la so hoan hao")
```
