bai 8 : Nhập số nguyên dương n. Lặp lại phép toán n = n + reverse(n) cho đến khi n là số đối xứng (palindrome). In ra số bước và giá trị palindrome thu được. 

```python
def is_palindrome(num):
    return str(num) == str(num)[::-1]
def reverse_num(num):
    return int(str(num)[::-1])
def lychrel_process(n):
    steps = 0
    while not is_palindrome(n):
        n = n + reverse_num(n)
        steps += 1
    return steps, n
n = int(input("Nhập số nguyên dương n: "))
steps, palindrome = lychrel_process(n)
print("Số bước:", steps)
print("Palindrome thu được:", palindrome)
```

bai 10 : Nhập vào số nguyên dương n. Tìm số x trong [1, n] có độ dài dãy Collatz lớn nhất và in ra x cùng độ dài dãy tương ứng (x L). Nếu có nhiều số cùng độ dài, chọn số nhỏ nhất.

```python
n = int(input())

max_length = 0
best_x = 0

for i in range(1, n + 1):
    current = i
    length = 1
    
    temp = current
    while temp != 1:
        if temp % 2 == 0:
            temp = temp // 2
        else:
            temp = 3 * temp + 1
        length += 1
    if length > max_length:
        max_length = length
        best_x = i

print(best_x, max_length)
```

bai 11 : 1 Viết chương trình đếm số lượng ước số chẵn của số nguyên dương n với n < 10^6.

```python
import math

n = int(input())
count = 0
for i in range(1, int(math.sqrt(n)) + 1):
    if n % i == 0:
        if i % 2 == 0:
            count += 1

        if (n // i) != i:
            if (n // i) % 2 == 0:
                count += 1

print(count)
```
