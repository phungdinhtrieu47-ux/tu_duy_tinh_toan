bai 3 : Tính tam giác: Nhâp vào ba số a,b,c bất kì. Kiểm tra xem nó có thể là 
độ dài ba cạnh hay không, nếu  không thì in  ra màn  hình  'Khong la ba  canh 
cua mot tam  giac'. Ngược lại, kiểm tra xem nó là tam giác gì: Đều, Cân, Vuông, hay Tam giác thường. Tính diện tích, chu vi của tam giác và in ra màn 
hình. 

```python
a, b, c =map(float,input("nhap vao ba canh cua tam giac : ").split())
if a+b > c and b+c >a and a+c > b:
    if a == b == c:
        print("day la tam giac deu")
    elif a == b or b == c or a == c:
        print("day la tam giac can")
    elif a**2 + b**2 == c**2 or b**2 + c**2 == a**2 or a**2 + c**2 == b**2:
        print(" day la tam giac vuong")
    else:
        print("day la tam giac thuong")
    p = (a + b + c) / 2
    s = (p * (p - a) * (p - b) * (p - c)) ** 0.5
    print(f"dien tich tam giac la : {s}")
    print(f"chu vi tam giac la : {a + b + c}")
else:
    print("day khong phai la ba canh cua tam giac")
```

bai 5 :  Viết chương trình giải phương trình bậc 2 (ax2 + bx + c = 0)

```python
a, b, c = map(float,input().split())
delta = b**2 - 4*a*c
if delta < 0:
    print("phuong trinh vo nghiem")
elif delta == 0:
    print(f"phuong trinh co nghiem kep x = {-b/(2*a)}")
else:
    x1 = (-b + delta**0.5) / (2*a)
    x2 = (-b - delta**0.5) / (2*a)
    print(f"phuong trinh co hai nghiem phan biet x1 = {x1} va x2 = {x2}")
```
