bai 15 : Nhập vào điểm trung bình của một học sinh và in ra học lực của học sinh đó. Xếp loại học lực theo quy tắc: - >= 8.0: Giỏi - >= 6.5: Khá - >= 5.0: Trung bình - < 5.0: Yếu

bai 16 :Nhập vào một số thực: in ra kết quả làm tròn lên ,tròn xuống, làm tròn tới số nguyên gần nhất (không được sử dụng hàm sẵn có của pyton). Các kết quả cách nhau một dấu cách 

```python
x = float(input())
floor_val = int(x) if x >= 0 or x == int(x) else int(x) - 1
ceil_val = int(x) if x == int(x) else (int(x) + 1 if x > 0 else int(x))
if x >= 0:
    nearest = int(x + 0.5)
else:
    nearest = int(x - 0.5)
print(ceil_val, floor_val, nearest)
```

bai 17 : Cho 4 số a, b, c, d. Hãy kiểm tra xem 4 số này có thể theo thứ tự tạo thành 1 cấp số nhân với công bội nguyên theo đúng thứ tự a, b, c, d hay không? Gợi ý : Tìm công bội (b / a) nếu b chia hết cho a, sau đó lấy b nhân công bội và so sánh vs c, c nhân công bội và so sánh vs 
d.

```python
a, b, c, d = map(int, input().split())
if a != 0 and b % a == 0:
    r = b // a
    if b * r == c and c * r == d:
        print("YES")
    else:
        print("NO")
else:
    print("NO")
```
