bai 16 : Viết hàm kiểm tra một số có phải số Amstronhg hay không? Hàm trả về kiểu bool, có tham số là một số nguyên x  

```python
def abc(x : int) -> bool:
    digits = str(x)
    n = len(digits)
    s = sum(int(digit) ** n for digit in digits)
    return s == x
print(abc(152))
```

bai 17 : Viết hàm kiêểm tra một số có phải số chính phương hay không? Hàm trả về kiểu bool, có tham số là một số nguyên x 

```python
def abc(x : int) -> bool:
    for i in range(1, int(x**0.5) + 1):
        if i * i == x:
            return True
    return False
print(abc(169))
```

bai 19 : VIết hàm kiểm tra một số có phải số kỳ diệu hay không, xử lý trên dữ liệu kiểu số nguyên. 

```python
def la_so_ky_dieu(n):
    n = abs(n)
    tong = 0
    while n > 0:
        tong += n % 10
        n //= 10
    return tong % 9 == 0
print(la_so_ky_dieu(18))
```

bai 20 : Viết hàm detect_number_type(s) trả về 'int' nếu s là số nguyên, 'float' nếu là số thực, ngược lại 'neither'. 

```python
def detect_number_type(n):
    if type(n) == int:
        return "Int"
    elif type(n) == float:
        return "Float"
    else:
        return "neither"
print(detect_number_type(5))
```

bai 23 :  Viết hàm quadratic_roots_count(a, b, c) trả về số nghiệm thực của phương trình a*x^2 + b*x + c = 0. 

```python
def quadratic_roots_count(a, b, c):
    discriminant = b**2 - 4*a*c
    if discriminant > 0:
        return 2
    elif discriminant == 0:
        return 1
    else:
        return 0
```

bai 24 :  Viết hàm are_collinear(x1,y1,x2,y2,x3,y3) kiểm tra ba điểm có thẳng hàng hay không. 

```python
def re_collinear(x1,y1,x2,y2,x3,y3):
    if (y2 - y1) * (x3 - x2) == (y3 - y2) * (x2 - x1):
        return True
    return False
print(re_collinear(1, 2, 2, 3, 3, 4))
```

bai 27 : Viết hàm trả về số lượng chữ số khác 0 trong n số nguyên nhập vào từ bàn phím (một list) 

```python
def count_nonzero_digits(nums):
    count = 0
    for n in nums:
        for digit in str(abs(n)):
            if digit != '0':
                count += 1
    return count
print(count_nonzero_digits([102, 305, -4000]))
```

bai 28 : Viết hàm tính góc ∠ABC (độ) từ tọa độ A(x1,y1), B(x2,y2), C(x3,y3).

```python
import math

def angle_ABC(A, B, C):
    x1, y1 = A
    x2, y2 = B
    x3, y3 = C
    BA = (x1 - x2, y1 - y2)
    BC = (x3 - x2, y3 - y2)
    dot = BA[0]*BC[0] + BA[1]*BC[1]
    len_BA = math.hypot(BA[0], BA[1])
    len_BC = math.hypot(BC[0], BC[1])
    if len_BA == 0 or len_BC == 0:
        return None
    cos_theta = dot / (len_BA * len_BC)
    cos_theta = max(-1, min(1, cos_theta))
    angle = math.degrees(math.acos(cos_theta))
    return angle
print(angle_ABC((0, 0), (1, 0), (1, 1)))
```

bai 30 : Viết hàm tính hệ số nhị thức C(n, k). 

```python
def C(n, k):
    if k < 0 or k > n:
        return 0
    k = min(k, n - k)
    result = 1
    for i in range(1, k + 1):
        result = result * (n - i + 1) // i
    
    return result
print(C(5, 2))
```

bai 31 : Viết hàm tính tổng bình phương các số từ 1 đến n.

bai 34 : Viết hàm sum_1_to_n(n) trả về tổng các số x từ 1 đến n, nếu x có các chữ số đôi một khác nhau Nếu n ≤ 0 trả về 0

```python
def sum_1_to_n(n):
    if n < 0:
        return 0
    tong = 0
    for i in range(1, n + 1):
        tong += i
        a = str(i)
        for b in range(1, len(a)):
            if a[b - 1] == a[b]:
                tong -= i
                break
    return tong
n = int(input())
print(sum_1_to_n(n))
            
        
```
