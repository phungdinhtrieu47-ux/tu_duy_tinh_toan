bai 18 :  Nhập vào số nguyên dương a và b, in toàn bộ ước chung của a và b 

```python
a , b = map(int, input().split())
if a > b:
    for i in range(1, b + 1):
        if a % i == 0 and b % i == 0:
            print(i, end=' ')
elif a < b:
    for i in range(1, a + 1):
        if a % i == 0 and b % i == 0:
            print(i, end=' ')
```

bai 25 : Nhập vào một dãy số nguyên, ngưng nhập khi người dùng nhập -1. Sau khi nhập xong, in số lớn nhất, số nhỏ nhất trong những số vừa nhập 

```python
first = True

while True:
    x = int(input())

    if x == -1:
        break

    if first:
        max_val = min_val = x
        first = False
    else:
        if x > max_val:
            max_val = x
        if x < min_val:
            min_val = x

if first:
    print("Khong co so nao duoc nhap")
else:
    print("So lon nhat:", max_val)
    print("So nho nhat:", min_val)
```

bai 26 : Dãy số fibonacci là dãy số được định nghĩa như sau: 1, 1, 2, 3, 5, 8, 13,... với số kế tiếp sẽ 
bằng tổng hai số trước đó Nhập vào số nguyên dương A, hãy tìm số trong dãy số fibonacci lớn nhất nhưng không vượt 
quá A

```python
a = int(input())
f1,f2 = 1,1
while f2 < a:
    f1, f2 = f2, f1 + f2
print(f1) 
```

bai 27 : Nhập vào một chuỗi, hãy đếm xem trong chuỗi có bao nhiêu từ (quy định là chuỗi không có 
ký tự đặc biệt, không số, không có dấu câu, chỉ có ký tự chữ và khoảng trắng) Dùng cách: for trên index; KHÔNG dùng hàm split(), len() 

```python
s = input()

dem = 0
i = 0

for _ in s:
    if s[i] != ' ' and (i == 0 or s[i - 1] == ' '):
        dem += 1
    i += 1

print(dem)
```

bai 29 : \Cho một chuỗi chỉ gồm các ký tự chữ cái thường và chữ cái hoa, hãy đổi ký tự hoa thành 
thường và thường thành hoa.

```python
s = input()

ket_qua = ""

for c in s:
    if c.islower():
        ket_qua += c.upper()
    else:
        ket_qua += c.lower()

print(ket_qua)
```

bai 30 : Nhập vào một chuỗi có dạng 3 số nguyên, mỗi số nguyên cách nhau một dấu phẩy, hãy tính

bai 31 : Nhập vào một chuỗi, hãy đếm xem trong chuỗi có bao nhiêu ký tự in hoa, bao nhiêu ký tự in 
thường, bao nhiêu ký tự số

```python
s = input()

hoa = 0
thuong = 0
so = 0

for c in s:
    if 'A' <= c <= 'Z':
        hoa += 1
    elif 'a' <= c <= 'z':
        thuong += 1
    elif '0' <= c <= '9':
        so += 1

print("Chu hoa:", hoa)
print("Chu thuong:", thuong)
print("Chu so:", so)
```
