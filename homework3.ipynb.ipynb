bai 9 :  Nhập vào số nguyên dương a, kiểm tra xem a có phải là số chính phương hay 
không? Số chính phương là số có căn bậc hai là một số nguyên VD: 1, 4, 9, 16, 25….

```python
import math
a = int(input())
x = int(math.sqrt(a))
if x * x == a:
    print(a, "la so chinh phuong")
else:
    print(a, "khong phai la so chinh phuong") 
```

bai 10 : Nhập vào số nguyên dương a, kiểm tra xem a có phải là số hoàn hảo hay không. Số hoàn hảo là số có tổng các ước (ước nhỏ hơn nó) bằng chính nó, VD: số 6, 28… 

```python
tong = 0
a = int(input())
for i in range(1, a):
    if a % i == 0:
        tong += i
if tong == a:
    print(a, "la so hoan hao")
else:
    print(a, "khong phai la so hoan hao")
```

bai 14 : Nhập vào một dãy số nguyên, dừng nhập khi người dùng nhập -1. Sau khi nhập xong, in số lớn nhất, số nhỏ nhất trong những số vừa nhập. KHÔNG ĐƯỢC dùng mảng

```python
first = True

while True:
    x = int(input())

    if x == -1:
        break

    if first:
        max_val = min_val = x
        first = False
    else:
        if x > max_val:
            max_val = x
        if x < min_val:
            min_val = x

if first:
    print("Khong co so nao duoc nhap")
else:
    print("So lon nhat:", max_val)
    print("So nho nhat:", min_val)
```

bai 16:  Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số chẵn, bao nhiêu chữ số lẻ 

```python
n = int(input())
chan = 0
le = 0
while n > 0:
    x = n % 10
    if x % 2 == 0:
        chan += 1
    else:
        le += 1
    n //=10
print("so chu so chan la:", chan)
print("so chu so le la:", le)
```

bai 19:Dãy số fibonacci là dãy số được định nghĩa như sau: 1, 1, 2, 3, 5, 8, 13,... với số kế tiếp sẽ bằng tổng hai số trước đó. Nhập vào số nguyên dương A, hãy tìm số trong dãy số fibonacci lớn nhất nhưng không vượt quá A

```python
a = int(input())
f1 = 1
f2 = 1
if a < 1:
    print("False")
else:
    while f2 <= a:
        f1, f2 = f2, f1 + f2
        
    print(f1)
```

bai 20 : Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số là số nguyên tố.

```python
n = int(input())
dem = 0
while n > 0:
    i = n % 10
    n //= 10
    if i < 2:
        dem += 0
    else:
        for j in range(2, int(i ** 0.5) + 1):
            if i % j == 0:
                break
        else:
                dem += 1
print(dem)
```
