bai 2 :  Một khu đất hình chữ nhật có chiều rộng a, chiều dài b. Người ra xây một khu vui chơi hình tròn tại vị trí trung tâm khu đất như hình vẽ. Nhập vào a, b. Hãy tính phần diện tích còn để trồng cây xung quanh là bao nhiêu. Biết pi = 3.14, lấy kết quả đến 2 chữ số phần thập phân.

```python
a, b = map(float, input("nhap chieu dai va chieu rong :").split())
s = a * b
S = 3.14 * (a / 2) ** 2
dien_tich_trong_cay = s - S
print("dien tich trong cay la :", dien_tich_trong_cay)
```

bai 6 : Nhập vào 3 số a, b, c. Kiểm tra xem đây có phải là cạnh của 1 tam giác không, nếu đúng thì tính diện tích của tam giác đó, in ra với 1 chữ số phần thập phân, nếu nó không phải 3 cạnh của tam giác thì in ra “Khong phai 3 canh cua tam giac”. 

```python
a, b, c = map(float, input("nhap vao ba canh cua tam giac :").split())
p = (a + b + c) / 2
s = (p * (p - a) * (p - b) * (p - c)) ** 0.5
if a + b > c and a + c > b and b + c > a:
    print("day la ba canh cua tam giac")
    print("dien tich cua tam giac la :", s)
else:
    print("khong phai 3 canh cua tam giac")
```
